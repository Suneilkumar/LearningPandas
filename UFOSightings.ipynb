```python
df.set_index('Date')
```

```python
df['MonthNo'] = df['Date'].apply(lambda x: x.month)
df.head(3)
```

```python
df['Day'] = df['Date'].apply(lambda x: x.day)
df['Year'] = df['Date'].apply(lambda x: x.year)
df['MonthName'] = df['Date'].apply(lambda x: x.strftime('%B'))
df['Weekday'] = df['Date'].apply(lambda x: x.strftime('%A'))
```

```python
df.head(10)
```

```python
p = df.groupby(['MonthName','Weekday']).size().unstack()
p.head(12)
```
